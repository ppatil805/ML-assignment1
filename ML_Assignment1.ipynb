```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv('/content/Linear Regression - Sheet1.csv')
df.head()
```

```python
df.info()
```

```python
plt.scatter(df.X,df.Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
clmn1=df[['X']]
clmn1[ :5]
```

```python
clmn2=df[['Y']]
clmn2[:5]
```

```python
model.fit(clmn1,clmn2)
```

```python
model.predict([[5]])
```

```python
model.predict([[9]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
test=df.sample(50)
test
```

```python
test1=test[['X']]
test1
test2=test[['Y']]
test2
```

```python
test_pred=model.predict(test1)
test_pred
```

```python
plt.scatter(df.X,df.Y,color='red',label='data points')
plt.plot(test1,test_pred,'k--',label='best fit line')
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
test2,test_pred
```

```python
mean_absolute_error(test2,test_pred)
```

```python
mean_squared_error(test2,test_pred)
```

```python
r2_score(test2,test_pred)
```
